```python
# Importing Libraries

import torch
from torchvision import transforms
from PIL import Image
```

```python
# Loading image

img = Image.open("/content/drive/MyDrive/Google Colab data/Homework 1/blue-anemone-flowers-close-up.jpeg")
img
```

```python
# Converting the image to a tensor

convert_tensor = transforms.ToTensor()
img_t = convert_tensor(img)
img_t
```

```python
# Calculating the mean of each channel

R_mean, G_mean ,B_mean = torch.mean(img_t, dim = [1,2])
```

```python
print("Mean across Read channel:", R_mean)
print("Mean across Green channel:", G_mean)
print("Mean across Blue channel:", B_mean)
```

```python
# Calculating the mean from tensor

Mean = torch.mean (img_t)
Mean
```
